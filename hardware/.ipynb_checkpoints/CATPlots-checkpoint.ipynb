# FieldFox CAT Plotter

```python
import matplotlib.pyplot as plt
import pandas as pd 
import os
```

```python
# 26
def plotCAT(filepath):
    df = pd.read_csv(filepath, 
                     skipinitialspace=True,
                     skiprows=26,
                     skipfooter=1,
                     engine='python',
                     header=None, 
                     names=["Hz", "dB"])
    display(df.dtypes)
    df["MHz"] = df["Hz"] / 1000_000
    
    df.plot(x="MHz", y="dB")
    
    plt.gca().invert_yaxis()

    peak = df.loc[df["dB"].idxmin()]
    plt.title(f"Insertion Loss - Peak: {peak['MHz']}MHz, {round(peak['dB'],2)}dB")
    plt.show()
```

```python
x_hx0057a_data = os.path.join("data", "x-hx0057a.csv")

plotCAT(x_hx0057a_data)
```

```python
# https://shop.kuhne-electronic.com/kuhne/en/shop/empfangsverstaerker/MKU+LNA+432+A+Selective+Low+Noise+Amplifier/?card=595#_tab_content5
kuhne_lna_data = os.path.join("data", "kuhne_lna_432_a.csv")

plotCAT(kuhne_lna_data)
```

```python
def plotNA(filepath, use_min=False):
    df = pd.read_csv(filepath, 
                     skipinitialspace=True,
                     skiprows=30,
                     skipfooter=1,
                     engine='python',
                     header=None, 
                     names=["Hz", "dB"])
    display(df.dtypes)
    df["MHz"] = df["Hz"] / 1000_000
    
    df.plot(x="MHz", y="dB")

    if use_min: 
        peak = df.loc[df["dB"].idxmin()]    
    else:
        peak = df.loc[df["dB"].idxmax()]
    plt.title(f"Network Analyzer - Peak: {peak['MHz']}MHz, {round(peak['dB'],2)}dB")
    plt.show()

bpf_data = os.path.join("data", "File_0BPF.csv")

plotNA(bpf_data)
```
